```python
import matplotlib.pyplot as plt
```

```python
string = 'Hola, mi nombre es Peyman y amo a Lana'
counter = dict()

for letter in string:
    if letter.isalpha():
        if letter in counter:
            counter[letter] += 1
        else:
            counter[letter] = 1

sorted_counter = sorted(counter.items(), key=lambda x: x[1], reverse=True)

letters = [item[0] for item in sorted_counter]
frequencies = [item[1] for item in sorted_counter]
```

```python
plt.bar(letters, frequencies)
plt.xlabel('Letters')
plt.ylabel('Frequencies')
plt.title('Letter Frequencies in the String')
plt.show()
```
