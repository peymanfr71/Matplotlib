```python
import matplotlib.pyplot as plt
```

```python
string = 'Hola, mi nombre es Peyman y amo a Lana'
counter = dict()
```

```python
for letter in string:
    if letter.isalpha():
        if letter in counter:
            counter[letter] += 1
        else:
            counter[letter] = 1

sorted_counter = sorted(counter.items(), key=lambda x: x[1], reverse=True)

letters = [item[0] for item in sorted_counter]
frequencies = [item[1] for item in sorted_counter]
```

```python
plt.pie(frequencies, labels=letters, autopct='%1.1f%%', startangle=140)
plt.title('Letter Frequencies as a Pie Chart')
plt.axis('equal')
plt.show()
```
