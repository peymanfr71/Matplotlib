```python
import matplotlib.pyplot as plt
import random
```

```python
string = 'Hola, mi nombre es Peyman y amo España, Japón, América.'
counter = dict()

for letter in string:
    if letter.isalpha():
        if letter in counter:
            counter[letter] += 1
        else:
            counter[letter] = 1

sorted_counter = sorted(counter.items(), key=lambda x: x[1], reverse=True)

letters = [item[0] for item in sorted_counter]
frequencies = [item[1] for item in sorted_counter]
```

```python
# Generate random colors for each bar
colors = ['#'+format(random.randint(0, 16777215), '06x') for _ in range(len(letters))]

plt.figure(figsize=(10, 6))
plt.bar(letters, frequencies, color=colors)
plt.xlabel('Letters')
plt.ylabel('Frequencies')
plt.title('Letter Frequencies in the String')
plt.xticks(rotation=45)
plt.tight_layout()

# Adding a legend
plt.legend(['Frequencies'], loc='upper right')
plt.show()
```
